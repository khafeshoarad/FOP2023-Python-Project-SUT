با استفاده از نمودار Pairplot برای هر دو جفت ستون موجود در دیتافریم یک نمودار پراکندگی به همراه KDE رسم کنید.
    </font>

```python
columns_for_pairplot = ['Area', 'Room', 'Parking', 'Warehouse', 'Elevator', 'Price', 'Price(usd)']

fig, axes = plt.subplots(nrows=len(columns_for_pairplot), ncols=len(columns_for_pairplot), figsize=(15, 15))

# Plot pair plots using subplots
for i in range(len(columns_for_pairplot)):
    for j in range(len(columns_for_pairplot)):
        if i != j:
            axes[i, j].scatter(df[columns_for_pairplot[j]], df[columns_for_pairplot[i]])
            axes[i, j].set_xlabel(columns_for_pairplot[j])
            axes[i, j].set_ylabel(columns_for_pairplot[i])
        else:
            axes[i, j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

نقشه حرارت یا Heatmap یک نوع نمودار تصویری است که اطلاعات را با استفاده از رنگ‌ها نشان می‌دهد. این نوع نمودار به شما این امکان را می‌دهد تا الگوها، تفاوت‌ها، و توزیع اطلاعات در یک مجموعه داده را به وضوح ببینید. کاربردهای Heatmap شامل موارد زیر می‌شوند:

    تجزیه و تحلیل داده‌های گسترده:
    Heatmap معمولاً برای نمایش داده‌های ماتریسی با ابعاد بالا استفاده می‌شود. این نمودار به تحلیل سریع و موثر داده‌های گسترده کمک می‌کند و به راحتی می‌توانید الگوها و تغییرات را تشخیص دهید.

    ارزیابی ماتریس ارتباط:
    Heatmap به شما این امکان را می‌دهد تا ارتباط بین داده‌ها را بصری تجسم کنید. می‌توانید ماتریس ارتباط بین متغیرها را با استفاده از رنگ‌ها نشان دهید، که باعث می‌شود الگوها به سرعت قابل درک باشند.

    شناسایی مقادیر بالا یا پایین:
    با استفاده از Heatmap، می‌توانید مقادیر بالا یا پایین در یک مجموعه داده را به سادگی شناسایی کنید. از رنگ‌های گرم و سرد برای نمایش مقادیر بالا و پایین استفاده می‌شود.

    نمایش توزیع مکانی داده‌ها:
    Heatmap می‌تواند به نمایش توزیع مکانی داده‌ها کمک کند. برای مثال، در حوزه GIS (سامانه اطلاعات جغرافیایی) می‌توان از Heatmap برای نمایش توزیع گرمایی نقاط جغرافیایی استفاده کرد.

    پیش‌بینی و پیش‌بینی:
    Heatmap به تحلیل الگوها و تغییرات در طول زمان کمک می‌کند. مثلاً در زمینه مالی، می‌توان از Heatmap برای نمایش نرخ سود و زیان در طول زمان استفاده کرد.

    پژوهش در حوزه پزشکی:
    در علوم پزشکی، Heatmap برای نمایش توزیع بیان ژن‌ها، نمایش توزیع گرمایی در تصاویر پزشکی، و یا تحلیل داده‌های مربوط به اثرات درمانی مورد استفاده قرار می‌گیرد.

در کل، Heatmap یک ابزار قدرتمند و بصری است که در تحلیل و نمایش داده‌ها برای درک بهتر الگوها و تغییرات بسیار موثر است    </font>

Pie chart یا نمودار دایره‌ای یک نوع نمودار است که برای نمایش توزیع نسبی اجزاء یک کل به کار می‌رود. این نمودار از یک دایره تشکیل شده است که به قطاع‌های مختلف تقسیم شده و هر قطاع نسبت بخشی از کل را نمایش می‌دهد. کاربردهای اصلی نمودار دایره‌ای شامل:

    نمایش توزیع درصدی:
    Pie chart به خوبی برای نمایش توزیع درصدی اجزاء یک کل مناسب است. با یک نگاه سریع به این نمودار، می‌توان فهمید که هر قسمت از دایره چه نسبتی از کل را نمایش می‌دهد.

    نمایش قسمت به کل:
    اگر شما به دنبال نشان دادن نسبت یک بخش به کل هستید، Pie chart یک ابزار مناسب است. این نمودار به سرعت اجزاء مختلف را در مقایسه با کل به نمایش می‌گذارد.

    تجسم اجزاء مختلف:
    اگر شما به دنبال تجسم اجزاء مختلف یک مجموعه هستید، Pie chart امکان مشاهده و مقایسه آسانی را فراهم می‌کند. این نمودار به راحتی قابل فهم است و اجازه می‌دهد تا اجزاء مختلف به صورت بصری مقایسه شوند.

    نمایش توزیع فراوانی:
    در برخی موارد، Pie chart ممکن است برای نمایش توزیع فراوانی یک متغیر مورد استفاده قرار گیرد. به عنوان مثال، اگر یک مجموعه داده را در چند دسته تقسیم کنید، Pie chart می‌تواند نشان دهنده توزیع فراوانی این دسته‌ها باشد.

    استفاده در گزارشات و ارائه‌ها:
    Pie chart به عنوان یک ابزار بصری مؤثر در گزارش‌ها، ارائه‌ها، و اسناد تجاری استفاده می‌شود. این نمودار به مخاطبان امکان می‌دهد به سرعت فهمیدن نسبت بین اجزاء و کل.

به طور کلی، Pie chart به عنوان یک ابزار ساده و مؤثر برای نمایش نسبت‌ها و توزیع اجزاء در یک کل مورد استفاده قرار می‌گیرد. هرچند که برخی از متخصصان آمار توصیه می‌کنند که از استفاده متعقدتر از Pie chart در مواردی که تعداد زیادی اجزاء داریم پرهیز کنیم و از نمودارهای دیگری مثل Bar chart یا Stacked bar chart استفاده کنیم    </font>

```python
df_sampled = df.sample(n=int(len(df) * .02)).drop(['Address'], axis=1)


sns.lineplot(data=df_sampled, x=df_sampled.index, y="Price(usd)", marker='o')


coefficients = np.polyfit(df_sampled['Area'], df_sampled['Price(usd)'], 1)
poly_function = np.poly1d(coefficients)

x_values = np.linspace(df_sampled['Area'].min(), df_sampled['Area'].max(), 1000)
y_values = poly_function(x_values)

sns.regplot(x = x_values, y=y_values, data=df_sampled)

plt.title('Line Plot of Price(usd)-Area')
plt.xlabel('Area')
plt.ylabel('Price(usd)')
plt.show()
```

نمودار میله‌ای یا Barplot یک نوع نمودار است که از میله‌ها (ستون‌ها) برای نمایش اطلاعات استفاده می‌کند. این نمودار برای نمایش توزیع یا مقایسه ارقام مختلف یک مجموعه داده بسیار مفید است. کاربردهای اصلی نمودار میله‌ای شامل:

    مقایسه مقادیر مختلف:
    با استفاده از Barplot، می‌توانید مقادیر مختلف یک متغیر را به راحتی با یکدیگر مقایسه کنید. این نمودار به سرعت اطلاعات را بصری و قابل فهم می‌کند.

    نمایش تغییرات در طول زمان:
    اگر می‌خواهید تغییرات یک متغیر یا مقدار در طول زمان را نشان دهید، می‌توانید از Barplot استفاده کنید. میله‌ها می‌توانند به عنوان دوره‌های زمانی یا دسته‌های زمانی مختلف عمل کنند.

    نمایش میانگین‌ها یا مجموع‌ها:
    Barplot برای نمایش میانگین یا مجموع مقادیر مختلف در دسته‌های مختلف کاربرد دارد. میله‌ها می‌توانند نمایانگر میانگین، مجموع، یا هر مقدار آماری دیگر باشند.

    نمایش توزیع فراوانی:
    اگر مجموعه داده شما شامل دسته‌های گوناگون است، Barplot به شما این امکان را می‌دهد که توزیع فراوانی این دسته‌ها را نمایش دهید.
